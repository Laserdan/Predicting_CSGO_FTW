#### **Team Weapon Values**

```python
df_file = dmg_demos[dmg_demos['file'] == 'esea_match_13770997.dem']

# En el pipeline cambiar el 'esea_match_13770997.dem' por el iterable
```

```python
df_file['round'].max()

# Este dato nos interesará de cara a hacer la iteración por todas las rondas
```

```python
df_round_ct = df_file[(df_file['round'] == 11) & (df_file['wp_type'] != 'Grenade') & (df_file['att_side'] == 'CounterTerrorist')]
df_round_t = df_file[(df_file['round'] == 11) & (df_file['wp_type'] != 'Grenade') & (df_file['att_side'] == 'Terrorist')]

# DF separado por equipos filtrando por nº ronda y armas diferentes a granadas

# En el pipeline cambiar el 11 por el iterable
```

```python
# wp_ct = test_df_round_ct[['att_id', 'wp']].set_index('att_id').T.to_dict('records')[0]
```

```python
wp_ct = df_round_ct[['att_id', 'wp']]
wp_t = df_round_t[['att_id', 'wp']]

wp_ct = wp_ct.drop_duplicates(subset='att_id')
wp_t = wp_t.drop_duplicates(subset='att_id')

# Drop_duplicates with subset to take the first record of each player. This way we reduce the error of taking a dropped gun
```

```python
print(wp_ct)
print(wp_t)
```

```python
wp_ct_val = sum([wp_values[weapon] for weapon in wp_ct['wp']])
wp_t_val = sum([wp_values[weapon] for weapon in wp_t['wp']])

# Value of the guns the team is carrying (CT & T)
```

```python
print(wp_ct_val)
print(wp_t_val)
```

```python
df_round_ct_nade = df_file_nade[(df_file_nade['round'] == 12) & (df_file_nade['att_side'] == 'CounterTerrorist')]
df_round_t_nade = df_file_nade[(df_file_nade['round'] == 11) & (df_file_nade['att_side'] == 'Terrorist')]

# DF separado por equipos filtrando por nº ronda

# En el pipeline cambiar el 11 por el iterable
```

```python
print(nade_ct_val)
print(nade_t_val)
```
